```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# إنشاء نسخة من البيانات
df_encoded = df.copy()

# إنشاء Label Encoder
le = LabelEncoder()

# عرض عدد القيم الفريدة لكل عمود
print("Unique Values لكل عمود:\n")
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].nunique()}")

print("\nEncoding...\n")

# تطبيق Label Encoding أو One-Hot Encoding
one_hot_cols = []

for col in df.select_dtypes(include='object').columns:
    if df[col].nunique() > 5:
        print(f"{col} --> Label Encoding")
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    else:
        print(f"{col} --> One-Hot Encoding")
        one_hot_cols.append(col)

# تطبيق One-Hot Encoding
df_encoded = pd.get_dummies(df_encoded, columns=one_hot_cols, drop_first=True)

# عرض أول 5 صفوف
df_encoded.head()
```
